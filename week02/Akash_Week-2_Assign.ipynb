```python
#http://www.gwicks.net/dictionaries.htm
english_words = []
with open('english3.txt','r') as file:
    for word in file:
        english_words.append(word.strip())
```

```python
def parsing(text):
    """
    This function performs text pre-processing
    and returns clean text
    """
    
    text = contractions.fix(text) # expands all the contractions in the text
    text = re.sub(r'[^\w\s]', '', text) # removes all the punctuation
    text = text.lower() # converts the text to lower case
    text = text.replace('\n',' ')  #replace newline character with empty string
    text = text.strip() # removes leading and trailing edge spaces of the string
    text = re.sub(' +',' ',text) # replaces more than one spaces with single space
    
    return text
```

```python
parsed_text1 = parsing(text1)
parsed_text2 = parsing(text2)
```

```python
def map_word_counts(text):
    """
    This functions returns tokenized words list from the given text 
    """
    words = text.split(' ')
    return words    
```

```python
def reducer(words_list):
    """
    This functions returns word counts for each unique word in the words list
    in the form of dictionary with key being word and value being counts.
    """
    unique, counts = np.unique(words_list, return_counts=True)
    word_counts = {}
    for i,j in zip(unique,counts):
        word_counts[i] = j
    return word_counts
```

```python
words = map_word_counts(parsed_text1)
words
```

```python
reducer(words)
```

```python
def map_non_english(text,english_words):
    """
    This functions returns tokenized non english words list from the given text 
    """
    words = text.split(' ')
    non_english_words = []
    for i in words:
        if i not in english_words:
            non_english_words.append(i)
        else:
            continue
    return non_english_words 
```

```python
non_english_words = map_non_english(parsed_text2,english_words)
non_english_words
```

```python
reducer(non_english_words)
```
